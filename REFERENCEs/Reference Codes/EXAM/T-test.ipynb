```python
import numpy as np
import pandas as pd
import math
from nltk import word_tokenize
from nltk.corpus import stopwords
```

```python
data=[]
with open("set1.txt","r") as p:
    x=p.read()
    data.extend(word_tokenize(x))
print(data)
stop_words=set(stopwords.words('english'))

t1=[]
for i in data:
    t1.append(i.lower())
#print(t1)
t=[]
for i in t1:
    if i not in stop_words:
        t.append(i)
#print(t)
token=[i for i in t if i.isalpha()]
print(token)

vocab=[]
for i in token:
    vocab.append(i)
#print(vocab)


```

```python
n=len(vocab)
cv=1.962
def t_test(w1,w2,w12,c1,c2,c12):
    mu=c1*c2/(n*n)
    xbar=c12/n
    t=(xbar-mu)/(math.sqrt(xbar/n))
    
    print(t)
    if(t>cv):
        print(w1+" "+w2+": "+"is a collocation")
    else:
        print(w1+" "+w2+": "+"not a collocation")

def chi_test(w1,w2,w12,c1,c2,c12):
    a=c12
    b=c1-a
    c=c2-a
    d=n-(a+b+c)
    e=a+b
    f=c+d
    g=b+d
    h=a+c

    e1=(e*h)/n
    e2=(e*g)/n
    e3=(f*h)/n
    e4=(f*g)/n

    s1=(a-e1)**2/e1
    s2=(b-e2)**2/e2
    s3=(c-e3)**2/e3
    s4=(d-e4)**2/e4
    chi=s1+s2+s3+s4
    if(chi>cv):
        print(w1+" "+w2+": "+"is a collocation")
    else:
        print(w1+" "+w2+": "+"not a collocation")
for i in range(len(vocab)-1):
    w1=vocab[i]
    w2=vocab[i+1]
    w12=w1+" "+w2
    c1=0
    c2=0
    c12=0
    for j in range(len(vocab)-1):
        if(vocab[j]==w1):
            c1+=1
            if(vocab[j+1]==w2):
                c12+=1
        if(vocab[j]==w2):
            c2+=1
    t_test(w1,w2,w12,c1,c2,c12)
    #chi_test(w1,w2,w12,c1,c2,c12)

```
