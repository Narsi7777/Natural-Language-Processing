# Viterbi Algoritham

```python
noStates=int(input("Enter No of states"))
states=[]
for i in range(noStates):
    states.append(input(f"Enter {i+1} th state"))
# print(states)
noObs=int(input("Enter No of Obervations"))
observations=[]
for i in range(noObs):
    observations.append(input(f"Enter {i+1} th Observation"))

transition_probability={}
for fromState in states:
    transition_probability[fromState]={}
    for toState in states:
        transition_probability[fromState][toState]=float(input(f"Enter Tranmiss Prob from {fromState} to {toState}  "))

        
        
emission_probability={}
for fromState in states:
    emission_probability[fromState]={}
    for toObs in observations:
        emission_probability[fromState][toObs]=float(input(f"Enter Emmission Prob from {fromState} to {toObs}  "))
        
```

```python
states = ('Healthy', 'Fever')
 
observations = ('normal', 'cold', 'dizzy')
 
start_probability = {'Healthy': 0.6, 'Fever': 0.4}
 
transition_probability = {
   'Healthy' : {'Healthy': 0.7, 'Fever': 0.3},
   'Fever' :   {'Healthy': 0.4, 'Fever': 0.6},
   }
 
emission_probability = {
   'Healthy' : {'normal': 0.5, 'cold': 0.4, 'dizzy': 0.1},
   'Fever'   : {'normal': 0.1, 'cold': 0.3, 'dizzy': 0.6},
   }
```

```python
def viterbi(obs,states,start_prob,trans_prob,emm_prob):
    path={s:[] for s in states}
    curr_prob={}
    for s in states:
        curr_prob[s]=round(start_prob[s]*emm_prob[s][obs[0]],5)
#     print(curr_prob)

    for i in range(1,len(obs)):
        print("Observation : ",obs[i])
        print("\n\n")
        last_prob=curr_prob
        curr_prob={}
        for curr_state in states:
            print("Current State : ",curr_state)
            print()
            max_prob=-1
            last_sta=""
            for last_state in states:
                #print(f"{last_state} To {curr_state}\n")
                x=last_prob[last_state]*trans_prob[last_state][curr_state]*emm_prob[curr_state][obs[i]]
                if(x>max_prob):
                    max_prob=x
                    last_sta=last_state
                print(f"{last_state} --> {curr_state} at {obs[i]} prob = {x}\n")
             #print("\n")
            curr_prob[curr_state]=max_prob
            path[curr_state].append(last_sta)
            print(f"Max Prob is {max_prob}\n\n\n")
    #print(path)
            
    #finding final largest probability
    max_prob=-1
    max_path=None
    for s in states:
        path[s].append(s)
        if(curr_prob[s]>max_prob):
            max_path=path[s]
            max_prob=curr_prob[s]

    print("Maxxx Path iss",max_path)


viterbi(observations,states,start_probability,transition_probability,emission_probability)
```
