```python
import pandas as pd
import math
from nltk.tokenize import word_tokenize
from nltk import bigrams

# Read the corpus from the file
df = pd.read_csv('Attachment ambiguity WITH.csv', encoding='latin1')
df.columns = ['data']
corpus = df['data']

# Combine all sentences into one string and tokenize
full_text = ' '.join(corpus.str.lower())
tokens = word_tokenize(full_text)

# Count the occurrences of elements and store in the dictionary
element_counts = {}
for element in tokens:
    element_counts[element] = element_counts.get(element, 0) + 1

# Generate bigrams from the tokenized words
bi_grams = list(bigrams(tokens))

# Count the frequency of each bigram
bigram_counts = {}
for bigram in bi_grams:
    bigram_counts[bigram] = bigram_counts.get(bigram, 0) + 1

# User input for noun, verb, and preposition
noun = input("Enter the Noun: ").strip().lower()
verb = input("Enter the Verb: ").strip().lower()
prep = input("Enter the Preposition: ").strip().lower()

# Get raw counts
n = element_counts.get(noun, 0)
v = element_counts.get(verb, 0)
p_n = bigram_counts.get((prep, noun), 0)
p_v = bigram_counts.get((prep, verb), 0)

print("p with n",p_n)
print("p_v",p_v)
# Calculate probabilities and lambda
def cal_prob(p_v, p_n, v, n): 
    prob_v = p_v / v if v > 0 else 0
    prob_n = p_n / n if n > 0 else 0
    return prob_v, prob_n

prob_v, prob_n = cal_prob(p_v, p_n, v, n)

def cal_lam(prob_v, prob_n):
    if prob_n == 0:
        return float('inf')  # Avoid divide-by-zero
    return math.log((prob_v * (1 - prob_n)) / prob_n, 2)

_lambda = cal_lam(prob_v, prob_n)

# Determine the attachment of the preposition
if _lambda > 0:
    print("The Preposition is attached with Verb.")
elif _lambda < 0:
    print("The Preposition is attached with Noun.")
else:
    print("The Preposition attachment cannot be determined.")
```

```python
import nltk
import pandas as pd
from nltk import word_tokenize, pos_tag
from collections import defaultdict
import math

# Read your dataset
df = pd.read_csv('Attachment ambiguity WITH.csv',encoding='latin1')  # Column: Sentence
df.columns=['Sentence']
sentences = df["Sentence"]  # List of sentences from the CSV

# Count pairs
verb_prep = defaultdict(int)  # Dictionary to count verb-preposition pairs
noun_prep = defaultdict(int)  # Dictionary to count noun-preposition pairs
prep_total = defaultdict(int)  # Dictionary to count total occurrences of each preposition

for sentence in sentences:  # Loop through each sentence in the dataset
    words = pos_tag(word_tokenize(sentence))  # Tokenize sentence and get POS tags (e.g., verb, noun, prep)
    
    # Loop through each word starting from the 3rd word (index 2)
    for i in range(2, len(words)):
        w, t = words[i]  # Current word and its POS tag
        pw, pt = words[i-1]  # Previous word and its POS tag
        ppw, ppt = words[i-2]  # Word before previous word and its POS tag
        
        # If the current word is a preposition (POS tag 'IN')
        if t == "IN":
            prep_total[w] += 1  # Increment total count of the preposition
            if pt.startswith("NN"):  # If the previous word is a noun
                noun_prep[(pw.lower(), w.lower())] += 1  # Count noun-prep pairs
            if ppt.startswith("VB"):  # If the word before the previous is a verb
                verb_prep[(ppw.lower(), w.lower())] += 1  # Count verb-prep pairs

# PMI function
def pmi(pair_count, total, prep_count, prep_total):
    if pair_count == 0 or prep_count == 0: return 0  # If counts are 0, return PMI as 0
    return math.log2((pair_count / total) / (prep_count / prep_total))  # PMI formula

# Predict attachment
def predict(v, n, p):
    tv, tn, tp = sum(verb_prep.values()), sum(noun_prep.values()), sum(prep_total.values())  # Total counts
    pv = pmi(verb_prep[(v, p)], tv, prep_total[p], tp)  # Compute PMI for verb-prep
    pn = pmi(noun_prep[(n, p)], tn, prep_total[p], tp)  # Compute PMI for noun-prep
    print("Attach to:", "VERB" if pv > pn else "NOUN")  # If PMI for verb is higher, attach to verb, else attach to noun

# EXAMPLE
predict("saw", "man", "with")  # Example: Predict whether "with" attaches to "saw" or "man"
```
